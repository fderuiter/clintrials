```python
from ggplot import ggplot, diamonds, aes, geom_density, scale_color_brewer, facet_wrap
```

```python
diamonds.head()
```

```python
%matplotlib inline
```

```python
ggplot(diamonds, aes(x='price', color='clarity')) + \
    geom_density() + \
    scale_color_brewer(type='div', palette=7) + \
    facet_wrap('cut')
```

The plots appear so ggplot is working.
