# WATU - Brock & Yap's Fusion Design

This IPython notebook illustrates the use of the `WATU` class in the `clintrials` package.

WATU fuses elements of Wages & Tait's efficacy/toxicity design with elements of Thall & Cook's EffTox utility contours.

```python
from clintrials import WATU
from clintrials.dosefinding.efftox import LpNormCurve
```

First, we define our efficacy skeletons, prior toxicity probabilities, and clinical cutoffs.

```python
prior_tox_probs = [0.05, 0.10, 0.15, 0.20, 0.25]
eff_skeletons = [
    [0.10, 0.20, 0.30, 0.40, 0.50],
    [0.20, 0.30, 0.40, 0.50, 0.50],
    [0.30, 0.40, 0.50, 0.50, 0.50]
]
tox_target = 0.25
tox_cutoff = 0.30
eff_cutoff = 0.20
trial_size = 30
first_dose = 1
```

Next, we define the utility contour using an `LpNormCurve`.

```python
metric = LpNormCurve(0.4, 0.7, 0.5, 0.4)
```

We instantiate the `WATU` object with our configured parameters:

```python
watu = WATU(
    skeletons=eff_skeletons,
    prior_tox_probs=prior_tox_probs,
    tox_target=tox_target,
    tox_limit=tox_cutoff,
    eff_limit=eff_cutoff,
    metric=metric,
    max_size=trial_size,
    first_dose=first_dose
)
```

Let's see the next dose to be administered according to the initial configuration.

```python
print("Initial next dose:", watu.next_dose())
```

We can update the model with clinical outcomes for an initial cohort of patients (dose, toxicity (1/0), efficacy (1/0)).

```python
watu.update([(1, 0, 0), (1, 0, 1), (1, 0, 0)])
print("Next dose after cohort 1:", watu.next_dose())
print("Recommended dose utilities:", watu.utility)
```
